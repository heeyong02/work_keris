```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel('학교별 교육편제단위 정보_20251002기준.xlsx', skiprows = 4,engine='openpyxl')
df['대계열분류'].unique()
```

```python
code = pd.read_excel('학교별 교육편제단위 정보_20251002기준.xlsx', sheet_name = 'Sheet2')
code['대계열분류'].nunique()
```

```python
code.columns
```

```python
import re

def clean_text(text):
    if pd.isna(text):
        return ''
    return re.sub(r'[^가-힣a-zA-Z0-9]', '', str(text))

df['대중소df'] = (df['대계열분류'].apply(clean_text) +
               df['중계열분류'].apply(clean_text) +
               df['소계열분류'].apply(clean_text))

code['대중소code'] = (code['대계열분류'].apply(clean_text) +
               code['중계열분류'].apply(clean_text) +
               code['소계열분류'].apply(clean_text))


df
```

```python
df = df.merge(
    code[['대중소code']],    
    left_on='대중소df',      
    right_on='대중소code',   
    how='left'
)
```
